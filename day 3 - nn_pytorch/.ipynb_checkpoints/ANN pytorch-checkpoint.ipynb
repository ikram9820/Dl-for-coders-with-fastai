```python
import torch
```

```python
def init_params(n_units):
    l = len(n_units)
    params = {}
    for i in range(1,l):
        params['w'+str(i)] = torch.randn([n_units[i] ,n_units[i-1]],dtype = torch.float32, requires_grad= True)
        params['b'+str(i)] = torch.zeros([n_units[i],1], dtype = torch.float32, requires_grad = True)
    return params
```

```python
def forward_prop(x,params):
    A = []
    Z = {}
    A.append(x)
    l = len(params)//2
    for i in range(1,l):
        w = params['w'+str(i)]
        b = params['b'+str(i)]
        print(len(A),i)
        Z[i] = torch.matmul( w, A[i-1] ) + b
        A.append( torch.clip(Z[i], 0.) )
  
            
    w = params['w'+str(l)]
    b = params['b'+str(l)]
    Z[l] = torch.matmul(w, A[l-1]) + b
#     A[l] = torch.sigmoid(Z[l])
    return Z[l]
    
```

```python
def compute_loss(pred , y):
    m = len(y)
    
    loss = (-1/m) * torch.sum( (torch.log(pred) * y) + (1 - torch.log(pred)) * (1-y) )
    return loss
```

```python
def optimize(loss, params, lr):
    loss.backward()
    l = len(params) //2 
    for i in (1,l):
        w = params['w'+str(i)]
        b = params['b'+str(i)]
        w.data -= w.grad.data *lr
        b.data -= b.grad.data *lr
        w.grad = None
        b.grad = None
        
    
```

```python
def model(x, n_units, lr):
    assert x.shape[0] == n_units[0] , 'the zeroth layer units must be equal to input shape[1] (2000,756)'
    
    params = init_params(n_units)
    pred = forward_prop(x,params)
    loss = compute_loss(pred,y)
    optimize(loss, params, lr)
    print(f'pred: {pred}, loss: {loss}')
```
