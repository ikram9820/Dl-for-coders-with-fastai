```python
import torch
```

```python
def init_params(n_units):
    l = len(n_units)
    params = {}
    for i in range(1,l):
        params['w'+str(i)] = torch.randn([n_units[i] ,n_units[i-1]],dtype = torch.float32, requires_grad= True)
        params['b'+str(i)] = torch.zeros([n_units[i],1], dtype = torch.float32, requires_grad = True)
    return params
```

```python
def forward_prop(x,params):
    A = []
    Z = {}
    A.append(x)
    l = len(params)//2
    for i in range(1,l):
        w = params['w'+str(i)]
        b = params['b'+str(i)]
        Z[i] = torch.matmul( w, A[i-1] ) + b
        A.append( torch.clip(Z[i], 0.) )
  
            
    w = params['w'+str(l)]
    b = params['b'+str(l)]
    Z[l] = torch.matmul(w, A[l-1]) + b
#     A[l] = torch.sigmoid(Z[l])
    return Z[l]
    
```

```python
def mse(pred , y):
    loss = ((pred - y )**2).mean()
    return loss
```

```python
def optimize(loss, params:dict, lr):
    loss.backward()
    for p in params:
        with torch.no_grad():
            params[p].data.sub_(params[p].grad.data *lr)
            params[p].grad = None
    
```
